```python
import re
#content = "i  am so affraid to see you"
#file = open("mark_speech.txt","r",encoding='utf-8')
file = open("Beckham_interview.txt","r",encoding='utf-8')
content = file.read()


#converting to lower case and removing punctuation
lower_content = content.lower()
punct_content = re.sub(r'[^\w\s]', '', lower_content)
```

```python
#tokenization + removing stop words + stemming + lemmatization
import Text_Prep as tp
#fin_content = tp.token(punct_content)
#print(fin_content)
#rm = tp.rm_stopwords(fin_content)
#print(rm)

transformed_content = tp.many_stem_lemma( tp.rm_stopwords(tp.token(punct_content)))
transformed_content
```

```python
#data downloading
import pandas as pd

data = pd.read_csv("Emotions_output1.csv")
```

```python
# data into dict
new_data = {}
for index,line in data.iterrows():
    word =line.values[1]
    emotion = line.values[2]
    new_data[word] = emotion

new_data
```

```python
#searching for existing words

emotions=[]
for x in transformed_content:
    for word in new_data:
        if word == x:
           emotions.append(new_data[word])
  
emotions       
     
```

```python
#counting emotions
from collections import Counter
emotions_frequency = Counter(emotions)
emotions_frequency.items()
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.bar(emotions_frequency.keys(),emotions_frequency.values())
plt.xlabel("Emotions")
plt.ylabel("Frequency")
plt.show()
```

```python
emotn =""
m = max(emotions_frequency.values())

for key in emotions_frequency:
    if emotions_frequency[key] == m:
        emotn = key

print("Le sentiment dominant est: ", emotn )
```
