```python
# Data downloading

import pandas as pd
csv_file = pd.read_csv('My_emotions1.csv',header=None)
```

```python
# Data into a dictionary

original_data = {}
for index,line in csv_file.iterrows():
    word =line.values[0]
    emotion = line.values[1]
    original_data[word] = emotion

len(original_data)
```

```python
# Data stemming and lemmatization

import Text_Prep as tp

new_data ={}
for key in original_data:
    new_word = tp.stem_lemma(key)
    new_data[new_word] = original_data[key]

new_data
```

```python
len(new_data)
```

```python
# dictionary to csv
df = pd.DataFrame(new_data.items(), columns = ['words','emotions'])

new_csv_file = 'Emotions_output1.csv'

df.to_csv(new_csv_file)
```
